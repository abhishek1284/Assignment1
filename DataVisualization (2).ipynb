```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Calculate mean, median, and standard deviation for 'Price' and 'Sales'
price_stats = {
    "mean": data_cleaned["Price"].mean(),
    "median": data_cleaned["Price"].median(),
    "std_dev": data_cleaned["Price"].std()
}

sales_stats = {
    "mean": data_cleaned["Sales"].mean(),
    "median": data_cleaned["Sales"].median(),
    "std_dev": data_cleaned["Sales"].std()
}

print("Price Statistics:", price_stats)
print("Sales Statistics:", sales_stats)
```

```python
# Calculate the correlation between 'Price' and 'Sales'
correlation = data_cleaned["Price"].corr(data_cleaned["Sales"])
print(f"Correlation between Price and Sales: {correlation}")

```

```python
correlation = data_cleaned["Price"].corr(data_cleaned["Sales"])
print(f"Correlation between Price and Sales: {correlation}")
if correlation > 0:
    print("Higher prices are associated with more sales.")
elif correlation < 0:
    print("Higher prices are associated with fewer sales.")
else:
    print("There is no correlation between Price and sales.")

```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data_cleaned["Price"], data_cleaned["Sales"], alpha=0.5, color='b')

# Add titles and labels
plt.title("Scatterplot of Price vs Sales", fontsize=14)
plt.xlabel("Price", fontsize=12)
plt.ylabel("Sales", fontsize=12)

# Show the plot
plt.show()
```

```python

# Ensure the 'Date' column is in datetime format
data_cleaned['Date'] = pd.to_datetime(data_cleaned['Date'])

# Sort the data by Date to ensure proper time order
data_cleaned = data_cleaned.sort_values(by='Date')

# Group the data by 'Date' and calculate the sum of sales per day
sales_by_date = data_cleaned.groupby('Date')['Sales'].sum()

# Create a line plot of Sales over time
plt.figure(figsize=(10, 6))
plt.plot(sales_by_date.index, sales_by_date.values, color='b', marker='o')

# Add titles and labels
plt.title("Sales Trend Over Time", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Sales", fontsize=12)

# Rotate date labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python

# Create a figure with two subplots for Price and Sales
plt.figure(figsize=(12, 6))

# Boxplot for Price
plt.subplot(1, 2, 1)
sns.boxplot(data=data_cleaned, x='Price', color='lightblue')
plt.title('Price Distribution')

# Boxplot for Sales
plt.subplot(1, 2, 2)
sns.boxplot(data=data_cleaned, x='Sales', color='lightgreen')
plt.title('Sales Distribution')

# Show the plots
plt.tight_layout()
plt.show()
```

```python
#To understand the overall trend of Sales when Price increases or decreases, we need to analyze how the two variables are related. Based on your previous request for correlation analysis (between Price and Sales), we can summarize the following points:

#Correlation: The positive correlation coefficient (0.66) indicates a moderate to strong positive relationship between Price and Sales in the dataset. This suggests that, in general, as Price increases, Sales also tend to increase.
```

```python
# Define price bins (adjust these ranges as necessary)
bins = [0, 100, 200, 300, 400, 500, 1000]  # Example price ranges
labels = ['0-100', '101-200', '201-300', '301-400', '401-500', '501+']  # Labels for the bins

# Create a new column for price range
data_cleaned['Price_Range'] = pd.cut(data_cleaned['Price'], bins=bins, labels=labels)

# Group by Price_Range and calculate the average sales for each range
sales_by_price_range = data_cleaned.groupby('Price_Range')['Sales'].mean()

# Display the results
print(sales_by_price_range)
```

```python

# Create the bar plot
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=sales_by_price_range.index, y=sales_by_price_range.values, color='lightblue')

# Add titles and labels
plt.title("Average Sales by Price Range", fontsize=14)
plt.xlabel("Price Range", fontsize=12)
plt.ylabel("Average Sales", fontsize=12)

# Display values on top of each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height():,.0f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                fontsize=12, color='black', 
                xytext=(0, 5), textcoords='offset points')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure 'Date' is in datetime format
data_cleaned['Date'] = pd.to_datetime(data_cleaned['Date'])

# Extract Month and Day of the Week from Date
data_cleaned['Month'] = data_cleaned['Date'].dt.month
data_cleaned['Day_of_Week'] = data_cleaned['Date'].dt.day_name()

# Group by Month and calculate the average sales for each month
sales_by_month = data_cleaned.groupby('Month')['Sales'].mean()

# Group by Day of the Week and calculate the average sales for each day
sales_by_day = data_cleaned.groupby('Day_of_Week')['Sales'].mean()

# Plotting the results
plt.figure(figsize=(12, 6))

# Plot sales by month
plt.subplot(1, 2, 1)
ax1 = sales_by_month.plot(kind='bar', color='lightblue')

# Set x-tick labels to month names
ax1.set_xticklabels(pd.to_datetime(sales_by_month.index, format='%m').month_name(), rotation=45)

plt.title('Average Sales by Month')
plt.xlabel('Month')
plt.ylabel('Average Sales')

# Plot sales by day of the week
plt.subplot(1, 2, 2)
sales_by_day = sales_by_day[['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']]  # Sort days in order
ax2 = sales_by_day.plot(kind='bar', color='lightgreen')
plt.title('Average Sales by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Sales')

# Show the plot
plt.tight_layout()
plt.show()
```
